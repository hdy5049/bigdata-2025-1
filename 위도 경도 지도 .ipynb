```python
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/자연휴양림출력화일.xlsx')
df
```

```python
anMap2 = folium.Map(location = [37.44791655386481, 127.16917766201597],
                zoom_start=15,
                zoom_control=False,
                control_scales=True
                )
anMap2
```

```python
from folium.plugins import MarkerCluster
mc = MarkerCluster().add_to(anMap2)

for n, row in df.iterrows():
  folium.Marker(location=[row['위도'], row['경도']], popup= folium.Popup(row['휴양림명'], max_width=200)).add_to(mc)

anMap2
```
